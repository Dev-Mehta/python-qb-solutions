````python
"""
You have been hired by an Airlines company to analyze their routes. The company has provided you following data.
Your task is to create a NetworkX directed graph representing the routes and display it.
Figure size should be (15,15), node color should be green, take appropriate node size, edge color should be red.
Data:
Kolkata to Mumbai
Mumbai to Pune
Mumbai to Goa
Kolkata to Delhi
Kolkata to Bhubaneshwar
Mumbai to Delhi
Delhi to Chandigarh
Delhi to Surat
Kolkata to Hyderabad
Hyderabad to Chennai
Chennai to Thiruvananthapuram
Thiruvananthapuram to Hyderabad
Kolkata to Varanasi
Delhi to Varanasi
Mumbai to Bangalore
Chennai to Bangalore
Hyderabad to Bangalore
Kolkata to Guwahati 
"""

g = nx.DiGraph()

init_text = """Kolkata to Mumbai
Mumbai to Pune
Mumbai to Goa
Kolkata to Delhi
Kolkata to Bhubaneshwar
Mumbai to Delhi
Delhi to Chandigarh
Delhi to Surat
Kolkata to Hyderabad
Hyderabad to Chennai
Chennai to Thiruvananthapuram
Thiruvananthapuram to Hyderabad
Kolkata to Varanasi
Delhi to Varanasi
Mumbai to Bangalore
Chennai to Bangalore
Hyderabad to Bangalore
Kolkata to Guwahati"""

# You don't need to write all cities one by one
# You can just split the text given in question by each line
routes = init_text.split('\n')
for route in routes:
    # Splitting each line by " to " will give you list like -> [city1, city2]. For eg: ["Kolkata", "Mumbai"]
    # We can directly add edges like this instead of writing route for each city one by one
    r = route.split(" to ")
    g.add_edges_from([(r[0], r[1])])
plt.figure(figsize=(15,15))
nx.draw_networkx(g, node_color='green',node_size=2000, font_size=6, edge_color='red')
````
